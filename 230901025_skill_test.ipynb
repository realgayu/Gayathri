```python
#1A
import statistics

#Sample exam scores (replace with your actual data)
exam_scores = [78, 65, 92, 88, 55, 72, 98, 60, 80, 75, 58, 82, 95, 70, 68, 85, 77, 63, 90, 89]

# 1. Overall Class Performance
average_score = statistics.mean(exam_scores)
highest_score = max(exam_scores)
lowest_score = min(exam_scores)

print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")

# 2. Pass/Fail Classification
pass_mark = 50
passed_students = [score for score in exam_scores if score >= pass_mark]
failed_students = [score for score in exam_scores if score < pass_mark]

print(f"\nNumber of students who passed: {len(passed_students)}")
print(f"Number of students who failed: {len(failed_students)}")

# 3. Grading on a Curve (example using fixed thresholds)
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score >= 50:
        return "E"
    else:
        return "F"

print("\nStudent Grades:")
for score in exam_scores:
    grade = get_grade(score)
    print(f"Score: {score}, Grade: {grade}")
```

```python
#1B
import numpy as np

#Input 2x2 matrix A
print("Enter the elements of 2x2 matrix A row-wise:")

A = []
for i in range(2):
    row = list(map(float, input(f"Row {i+1} (enter 2 numbers separated by space): ").split()))
    if len(row) != 2:
        raise ValueError("Each row must contain exactly 2 numbers.")
    A.append(row)

A = np.array(A)

#Input 2x1 vector b
print("Enter the elements of 2x1 vector b:")
b = []
for i in range(2):
    val = float(input(f"Element {i+1}: "))
    b.append(val)

b = np.array(b)

try:
    x = np.linalg.solve(A, b)
    print("\nThe solution vector x is:")
    print(x)
except np.linalg.LinAlgError:
    print("\nThe system has no unique solution (matrix might be singular).")
```

```python
#2B
import numpy as np
import matplotlib.pyplot as plt

print("Enter the grades of 30 students for Math, Science, and English.")
print("Each entry should be 3 space-separated integers (e.g., 85 90 78)")

grades = []
for i in range(30):
    while True:
        try:
            entry = list(map(int, input(f"Student {i+1} grades: ").split()))
            if len(entry) != 3 or not all(0 <= x <= 100 for x in entry):
                raise ValueError
            grades.append(entry)
            break
        except ValueError:
            print("Invalid input. Please enter 3 integers between 0 and 100.")

grades = np.array(grades)  # Shape: (30, 3)
subjects = ['Math', 'Science', 'English']


print("\nSubject-wise Statistics:")
for i, subject in enumerate(subjects):
    subject_grades = grades[:, i]
    print(f"{subject}:")
    print(f"  Average: {np.mean(subject_grades):.2f}")
    print(f"  Highest: {np.max(subject_grades)}")
    print(f"  Lowest : {np.min(subject_grades)}")


total_scores = np.sum(grades, axis=1)
average_scores = np.mean(grades, axis=1)

top_index = np.argmax(total_scores)
print("\nTop Performing Student:")
print(f"  Student {top_index+1} -> Total: {total_scores[top_index]}, Average: {average_scores[top_index]:.2f}")

plt.figure(figsize=(12, 5))

for i, subject in enumerate(subjects):
    plt.subplot(1, 3, i+1)
    plt.hist(grades[:, i], bins=10, edgecolor='black', alpha=0.7)
    plt.title(f'{subject} Grade Distribution')
    plt.xlabel('Grade')
    plt.ylabel('Number of Students')
    plt.grid(True)

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
avg_subject_scores = np.mean(grades, axis=0)
plt.bar(subjects, avg_subject_scores, color=['blue', 'green', 'orange'])
plt.title('Average Score per Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#3A
import cmath

def get_impedance(component_type, value, frequency):
    omega = 2 * cmath.pi * frequency
    if component_type == 'R':
        return complex(value)
    elif component_type == 'C':
        return complex(0, -1 / (omega * value))
    elif component_type == 'L':
        return complex(0, omega * value)
    else:
        raise ValueError("Invalid component type. Must be 'R', 'L', or 'C'.")

def compute_total_impedance(impedance_list, configuration):
    if configuration == 'series':
        return sum(impedance_list)
    elif configuration == 'parallel':
        inv_total = sum(1 / z for z in impedance_list)
        return 1 / inv_total if inv_total != 0 else complex('inf')
    else:
        raise ValueError("Invalid configuration. Must be 'series' or 'parallel'.")

def main():
    print("=== Total Impedance Calculator ===")
    
    try:
        frequency = float(input("Enter the frequency of the AC source (in Hz): "))
        num_components = int(input("Enter the number of components: "))
        
        components = []
        for i in range(num_components):
            ctype = input(f"Enter type of component #{i+1} (R/L/C): ").strip().upper()
            value = float(input(f"Enter the value (Ohms for R, Henry for L, Farads for C): "))
            impedance = get_impedance(ctype, value, frequency)
            components.append(impedance)
            print(f"  Impedance of component #{i+1} ({ctype}) = {impedance:.2f} Ω")
        
        config = input("Enter the configuration (series/parallel): ").strip().lower()
        total = compute_total_impedance(components, config)
        
        print(f"\nTotal impedance of the circuit ({config}): {total:.2f} Ω")
    except Exception as e:
        print("Error:", e)

if _name_ == "_main_":
    main()
```

```python
#3B
import numpy as np

print("Enter daily price data for STOCK A (Open, High, Low) for 5 days:")
stock_a = []
for i in range(5):
    while True:
        try:
            row = list(map(float, input(f"Day {i+1} (Open High Low): ").split()))
            if len(row) != 3:
                raise ValueError
            stock_a.append(row)
            break
        except ValueError:
            print("Invalid input. Please enter 3 numbers (Open High Low).")

print("\nEnter daily price data for STOCK B (Open, High, Low) for 5 days:")
stock_b = []
for i in range(5):
    while True:
        try:
            row = list(map(float, input(f"Day {i+1} (Open High Low): ").split()))
            if len(row) != 3:
                raise ValueError
            stock_b.append(row)
            break
        except ValueError:
            print("Invalid input. Please enter 3 numbers (Open High Low).")

stock_a = np.array(stock_a)
stock_b = np.array(stock_b)


sum_matrix = stock_a + stock_b
diff_matrix = stock_a - stock_b


print("\nElement-wise Sum of Stock A and B (Combined Price Values):")
print(sum_matrix)

print("\nElement-wise Difference of Stock A and B (Daily Spread):")
print(diff_matrix)
```

```python
#4A
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = input("Enter the path to the EEG CSV data file: ")
df = pd.read_csv(file_path)

print("\nFirst few rows of the data:")
print(df.head())


participants = {}
for col in df.columns:
    if '_' in col:
        pid, eid = col.split('_')
        if pid not in participants:
            participants[pid] = []
        participants[pid].append(col)

print("\n--- Statistical Analysis ---")
for pid, electrodes in participants.items():
    print(f"\nParticipant: {pid}")
    for electrode in electrodes:
        signal = df[electrode]
        print(f"  Electrode {electrode}:")
        print(f"    Mean     : {signal.mean():.3f}")
        print(f"    Variance : {signal.var():.3f}")
        print(f"    Peak (Max): {signal.max():.3f}")

plot_choice = input("\nWould you like to visualize brainwave activity? (y/n): ").strip().lower()
if plot_choice == 'y':
    for pid, electrodes in participants.items():
        plt.figure(figsize=(10, 6))
        for electrode in electrodes:
            plt.plot(df[electrode], label=electrode)
        plt.title(f'EEG Signals for {pid}')
        plt.xlabel('Time Points')
        plt.ylabel('Brainwave Activity')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

```python
#4B
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

df = pd.DataFrame(data)


print("1. Inventory Data:")
print(df)

df['Total_Value'] = df['Price'] * df['Quantity']
print("\n2. Data with Total Value:")
print(df)


most_expensive = df.loc[df['Price'].idxmax()]
print("\n3. Most Expensive Product:")
print(most_expensive)

total_inventory_value = df['Total_Value'].sum()
print("\n4. Total Inventory Value:", total_inventory_value)


high_stock = df[df['Quantity'] > 20]
print("\n5. Products with quantity > 20:")
print(high_stock)


plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Total_Value'], color='skyblue')
plt.title('Total Value of Inventory per Product')
plt.xlabel('Product')
plt.ylabel('Total Value ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
