```python
#Line Plot with Title and Labels
import matplotlib.pyplot as plt
import numpy as np
#Title and label
ypoints = np.array([3, 8, 1, 10, 5, 7])
print(plt.plot(ypoints, linestyle='dotted'))
plt.title('Sample graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
print(plt.show())
```
